```python
# Import libraries
import pandas as pd
import numpy as np
import warnings 
from datetime import datetime
warnings.filterwarnings ('ignore')
```

```python
# Dataset download
hour_pollullant = pd.read_csv ('export to tableau.csv')
```

```python
# Supongamos que el formato actual de 'DATE' es '11/10/2023' (día/mes/año)
hour_pollullant['DATE'] = pd.to_datetime(hour_pollullant['DATE'], format='%Y/%m/%d').dt.strftime('%Y-%m-%d')
```

```python
hour_pollullant.DATE.info ()
```

```python
import pandas as pd

# Supongamos que 'hour_pollutant' es tu DataFrame

# Identificar las columnas que deben permanecer intactas
columnas_fijas = ['DISTRICT', 'DATE', 'POLLULLANT', 'TYPE', 'URBAN_AREA', 'RATIO', 'ID', 'Status']

# Derretir el DataFrame
hour_pollutant_melted = pd.melt(hour_pollullant, 
                                id_vars=columnas_fijas, 
                                var_name='HOUR', 
                                value_name='VALUE')

# Combina DATE y HOUR y convierte '24' a '00' en la siguiente fecha
hour_pollutant_melted['DATETIME'] = (hour_pollutant_melted['DATE'] + ' ' + 
                                     hour_pollutant_melted['HOUR'].replace('24', '00') + ':00:00')

# Convierte la columna DATETIME a formato datetime
hour_pollutant_melted['DATETIME'] = pd.to_datetime(hour_pollutant_melted['DATETIME'], format='%Y-%m-%d %H:%M:%S')

# Elimina las columnas DATE y HOUR
hour_pollutant_melted = hour_pollutant_melted.drop(['DATE', 'HOUR'], axis=1)

# Reordena las columnas
hour_pollutant_melted = hour_pollutant_melted[['DISTRICT', 'DATETIME', 'POLLULLANT', 'TYPE', 'URBAN_AREA', 'VALUE', 'RATIO', 'ID', 'Status']]

# Muestra el DataFrame final
print(hour_pollutant_melted)
```
